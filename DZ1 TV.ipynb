# –£—Ä–æ–∫ 1
–°–ª—É—á–∞–π–Ω—ã–µ —Å–æ–±—ã—Ç–∏—è. –£—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å. –§–æ—Ä–º—É–ª–∞ –ë–∞–π–µ—Å–∞. –ù–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –∏—Å–ø—ã—Ç–∞–Ω–∏—è
–î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ

```python
import numpy as np
```

–ó–∞–¥–∞—á–∞ 1

–ò–∑ –∫–æ–ª–æ–¥—ã –≤ 52 –∫–∞—Ä—Ç—ã –≤—ã–Ω–∏–º–∞—é—Ç —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º 4 –∫–∞—Ä—Ç—ã. –ù–∞–π—Ç–∏ —á–∏—Å–ª–æ –∏—Å—Ö–æ–¥–æ–≤, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ç–æ–º—É, —á—Ç–æ –±—ã–ª –≤—ã—Ç—è–Ω—É—Ç —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç—É–∑.

```python
# –°–Ω–∞—á–∞–ª–∞ –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ (—Å–æ—á–µ—Ç–∞–Ω–∏–π), –∫–æ—Ç–æ—Ä—ã–º–∏ –º–æ–∂–Ω–æ –≤—ã—Ç–∞—â–∏—Ç—å 4 –∫–∞—Ä—Ç—ã: 
def combinations(n: int, k: int) -> int:
    """–ß–∏—Å–ª–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π"""
    return np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k))
fours = combinations(52, 4)
print(fours)
```

```python
# "–•–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç—É–∑" –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤ –≤—ã–Ω—É—Ç—ã—Ö –∫–∞—Ä—Ç–∞—Ö –º–æ–∂–µ—Ç –±—ã—Ç—å 1, 2, 3 –∏–ª–∏ 4 —Ç—É–∑–∞.
# –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–∞–∂–¥—ã–π —Å–ª—É—á–∞–π –æ—Ç–¥–µ–ª—å–Ω–æ:
t1 = combinations(4, 1)*combinations(48, 3)
t2 = combinations(4, 2)*combinations(48, 2)
t3 = combinations(4, 3)*combinations(48, 1)
t4 = 1
print(t1+t2+t3+t4)
```

```python
# –£–¥–æ–±–Ω–µ–µ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å —á–∏—Å–ª–æ –ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤ –∏ –æ—Ç–Ω—è—Ç—å –∏—Ö –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞.
# –ò–∑ 48 –∫–∞—Ä—Ç, –Ω–µ —è–≤–ª—è—é—â–∏—Ö—Å—è —Ç—É–∑–∞–º–∏, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –≤—ã–Ω—É—Ç—å 4 –∫–∞—Ä—Ç—ã:
not_t = combinations(48, 4)
print(not_t)
```

```python
# –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —á–∏—Å–ª–æ –∏—Å—Ö–æ–¥–æ–≤, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ç–æ–º—É, —á—Ç–æ –±—ã–ª –≤—ã—Ç—è–Ω—É—Ç —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ç—É–∑:
t = fours - not_t
print(t)
```

–ó–∞–¥–∞—á–∞ 2

–°–µ–º—å —á–µ–ª–æ–≤–µ–∫ —Ä–∞—Å—Å–∞–∂–∏–≤–∞—é—Ç—Å—è –Ω–∞—É–¥–∞—á—É –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Ä—è–¥–æ–º –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å: –∞) –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞? –±) —Ç—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞?

```python
def permutations(n: int) -> int:
    """–ß–∏—Å–ª–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫"""
    return np.math.factorial(n)
# –ü–æ—Å—á–∏—Ç–∞–µ–º —á–∏—Å–ª–æ —Å–ø–æ—Å–æ–±–æ–≤ —Å–µ—Å—Ç—å –¥–ª—è —Å–µ–º–∏ —á–µ–ª–æ–≤–µ–∫:
n = permutations(7)
print(n)
```

```python
# –î–ª—è —Å–µ–º–∏ —á–µ–ª–æ–≤–µ–∫ –≤—Å–µ–≥–æ –ø–∞—Ä —Å–æ—Å–µ–¥–Ω–∏—Ö –º–µ—Å—Ç - 6. 
# –ö–∞–∂–¥–∞—è –ø–∞—Ä–∞ —á–µ–ª–æ–≤–µ–∫ –º–æ–∂–µ—Ç —Å–µ—Å—Ç—å –Ω–∞ —ç—Ç–∏ –º–µ—Å—Ç–∞ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏, –∞ –æ—Å—Ç–∞–ª—å–Ω—ã–µ 5 –ª—é–±—ã–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏ (–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏, 5!).
# –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —á–∏—Å–ª–æ —Å–ø–æ—Å–æ–±–æ–≤ —Å–µ—Å—Ç—å —Ä—è–¥–æ–º –¥–ª—è –¥–≤—É—Ö –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö –ª—é–¥–µ–π:
m2 = 6*2*permutations(5)
# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º:
print(m2/n)  # —ç—Ç–æ 2/7
```

```python
# –¢–∞–∫ –∫–∞–∫ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ä–∞—Å—Å–∞–¥–∫–∏ –≤–Ω—É—Ç—Ä–∏ —Ç—Ä–æ–π–∫–∏ –≤ —É—Å–ª–æ–≤–∏–∏ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞, —Ç–æ –ø–æ –∞–Ω–∞–ª–æ–≥–∏–∏:
# —Ç—Ä–æ–µ–∫ –º–µ—Å—Ç —Ä—è–¥–æ–º –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ 5,
# –∫–∞–∂–¥—ã–µ —Ç—Ä–∏ —á–µ–ª–æ–≤–µ–∫–∞ –º–æ–≥—É—Ç —Å–µ—Å—Ç—å –Ω–∞ –º–µ—Å—Ç–∞ —à–µ—Å—Ç—å—é (3!) —Å–ø–æ—Å–æ–±–∞–º–∏, –∞ –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ 24 —Å–ø–æ—Å–æ–±–∞–º–∏ (4!).
m3 = 5*6*24
# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Ç—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º:
print(m3/n) # —ç—Ç–æ 1/7
```

        –∞) –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞: 2/7 –∏–ª–∏ 0.2857142857142857
        –±) —Ç—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞: 1/7 –∏–ª–∏ 0.14285714285714285

```python
# –î–ª—è –ø–µ—Ä–≤–æ–≥–æ –≤–æ–ø—Ä–æ—Å–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –µ–≥–æ –∑–Ω–∞–µ—Ç, 5/6.
# –î–ª—è –≤—Ç–æ—Ä–æ–≥–æ, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –ø–µ—Ä–≤—ã–π: 49/59
# –î–ª—è —Ç—Ä–µ—Ç—å–≥–æ, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –ø–µ—Ä–≤—ã–π –∏ –≤—Ç–æ—Ä–æ–π: 48/58
```

```python
#  –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ —Ç—Ä–∏ –≤–æ–ø—Ä–æ—Å–∞:
p3 = (5*49*48)/(6*59*58)
print(p3)
```

```python
#  –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –¥–≤–∞ –≤–æ–ø—Ä–æ—Å–∞:
p2 = (5*49)/(6*59)
print(p2)
```

    –∞) –≤—Å–µ: 0.5727644652250146

–ë—Ä–æ—Å–∞–µ—Ç—Å—è –∏–≥—Ä–∞–ª—å–Ω–∞—è –∫–æ—Å—Ç—å. –ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ A - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á—ë—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞, —Å–æ–±—ã—Ç–∏–µ B - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ –±–æ–ª—å—à–µ —Ç—Ä—ë—Ö. –Ø–≤–ª—è—é—Ç—Å—è –ª–∏ —ç—Ç–∏ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏?

```python
# –°—á–∏—Ç–∞–µ–º, —á—Ç–æ ùëÉ(ùê¥)=ùëÉ(ùêµ)=1/2. –ü–æ—Å—á–∏—Ç–∞–µ–º —Å–æ–≤–º–µ—Å—Ç–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(AB). –°–æ–±—ã—Ç–∏–µ ùê¥ùêµ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ø–æ—è–≤–ª–µ–Ω–∏—é —á–∏—Å–µ–ª 4 –∏ 6,
# —Ç.–µ. P(AB)=1/3
# –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, 
# ùëÉ(ùê¥|ùêµ) = P(AB)/ùëÉ(ùêµ) = (1/3)/(1/2) = 2/3
# –ü–æ—Å–∫–æ–ª—å–∫—É ùëÉ(ùê¥|ùêµ) –Ω–µ —Ä–∞–≤–Ω–æ ùëÉ(ùê¥), –∑–∞–∫–ª—é—á–∞–µ–º, —á—Ç–æ —Å–æ–±—ã—Ç–∏—è A –∏ B –∑–∞–≤–∏—Å–∏–º—ã.
```

–ó–∞–¥–∞—á–∞ 5

–î–æ–ø—É—Å—Ç–∏–º, –∏–º–µ–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–∞—è –æ—á–µ–Ω—å —Ä–µ–¥–∫–∞—è –±–æ–ª–µ–∑–Ω—å (–ø–æ—Ä–∞–∂–∞–µ—Ç 0.1% –Ω–∞—Å–µ–ª–µ–Ω–∏—è). –í—ã –ø—Ä–∏—Ö–æ–¥–∏—Ç–µ –∫ –≤—Ä–∞—á—É, –≤–∞–º –¥–µ–ª–∞—é—Ç —Ç–µ—Å—Ç –Ω–∞ —ç—Ç—É –±–æ–ª–µ–∑–Ω—å, –∏ —Ç–µ—Å—Ç –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º. –í—Ä–∞—á –≥–æ–≤–æ—Ä–∏—Ç –≤–∞–º, —á—Ç–æ —ç—Ç–æ—Ç —Ç–µ—Å—Ç –≤–µ—Ä–Ω–æ –≤—ã—è–≤–ª—è–µ—Ç 99% –±–æ–ª—å–Ω—ã—Ö —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é –∏ –≤—Å–µ–≥–æ –ª–∏—à—å –≤ 1% —Å–ª—É—á–∞–µ–≤ –¥–∞—ë—Ç –ª–æ–∂–Ω—ã–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç.

–í–æ–ø—Ä–æ—Å: –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã –µ–π?

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤–Ω–æ–≤—å –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ —Å —Ä–∞—Å–∫—Ä—ã—Ç–∏–µ–º –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º—É–ª—ã –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

```python
# –ü—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º —Ç–µ—Å—Ç–µ –≤–æ–∑–º–æ–∂–Ω—ã —Ç—Ä–∏ –≤–∞—Ä–∏–∞–Ω—Ç–∞:
# 1) –≤—ã –≤—Ö–æ–¥–∏—Ç–µ –≤ –≥—Ä—É–ø–ø—É —Ç–µ—Ö, –∫–æ–≥–æ –ø–æ—Ä–∞–∂–∞–µ—Ç –±–æ–ª–µ–∑–Ω—å, –∏ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ 0,001*0,99 = 0,00099
# 2) –≤—ã –≤—Ö–æ–¥–∏—Ç–µ –≤ –≥—Ä—É–ø–ø—É —Ç–µ—Ö, –∫–æ–≥–æ –ø–æ—Ä–∞–∂–∞–µ—Ç –±–æ–ª–µ–∑–Ω—å, –Ω–æ –≤—ã –Ω–µ –±–æ–ª—å–Ω—ã, —Ç–µ—Å—Ç –ª–æ–∂–Ω—ã–π. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ 0,001*0,01 = 0,00001
# 3) –≤—ã –Ω–µ –≤—Ö–æ–¥–∏—Ç–µ –≤ –≥—Ä—É–ø–ø—É —Ç–µ—Ö, –∫–æ–≥–æ –ø–æ—Ä–∞–∂–∞–µ—Ç –±–æ–ª–µ–∑–Ω—å, –∏ –≤—ã –Ω–µ –±–æ–ª—å–Ω—ã, —Ç–µ—Å—Ç –ª–æ–∂–Ω—ã–π. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ 0,999*0,01 = 0,00999
```

```python
# –í—ã—á–∏—Å–ª–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∞—Ä–∏–∞–Ω—Ç–∞ 1, —Ç.–µ, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã –ø—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º —Ç–µ—Å—Ç–µ:
p = 0.00099/(0.00099+0.00001+0.00999)
print(p)
```
